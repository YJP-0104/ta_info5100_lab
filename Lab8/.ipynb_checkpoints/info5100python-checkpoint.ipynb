```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
video_game_sale = pd.read_csv('vgsales.csv')
```

```python
video_game_sale["NA_Sales"].value_counts()
```

```python
plt.figure(figsize=(15, 6))  
plt.plot(video_game_sale['Name'], video_game_sale['NA_Sales'], color='skyblue')  
plt.title("North America Sales of Video Games", fontsize=14, fontweight='bold')
plt.xlabel("Game Names", fontsize=12)
plt.ylabel("Sales in North America (in millions)", fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10) 
plt.tight_layout() 

# Save chart as an image file
plt.savefig("North_America_Video_Game_Sales.png", dpi=300)
plt.show()
```

```python
plt.grid(True)

plt.plot(video_game_sale["Name"], video_game_sale['NA_Sales'])
plt.title("Video Game Sales in North America")
plt.xlabel("Name")
plt.ylabel("NA_Salse")
# Set x-axis for every 10 units
plt.xticks(range(0, max(video_game_sale["Rank"]) + 1, 10))  
plt.show()
```

```python
sns.histplot(data=video_game_sale, x="Genre")
plt.title("Sales Distribution")
plt.xlabel("Genre")
plt.ylabel("Count")
```

```python
sns.pairplot(video_game_sale, hue="Genre")
plt.show()
```

```python
# Create a KDE plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=video_game_sale, x="Global_Sales", hue="Genre", fill=True, common_norm=False, palette="crest", alpha=0.2)

# Add labels and title
plt.title("Sales based on different Genre")
plt.xlabel("Genre")
plt.ylabel("Sales")
plt.show()
```
