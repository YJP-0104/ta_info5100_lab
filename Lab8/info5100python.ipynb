```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
video_game_sale = pd.read_csv('vgsales.csv')
```

```python
video_game_sale["NA_Sales"].value_counts()
```

```python
plt.plot(video_game_sale['Name'], video_game_sale['NA_Sales'])
plt.title("Games Sales")
plt.xlabel("Name")
plt.ylabel("North America Sales")
plt.xticks(rotation=90)  
plt.tight_layout()  
plt.savefig("game_sales_chart.png", dpi=300)  
plt.show()
```

```python
plt.grid(True)

plt.plot(video_game_sale["Name"], video_game_sale['NA_Sales'])
plt.title("Video Game Sales in North America")
plt.xlabel("Name")
plt.ylabel("NA_Salse")
# Set x-axis for every 10 units
plt.xticks(range(0, max(video_game_sale["Rank"]) + 1, 10))  
plt.savefig("GameSales.png", dpi=300)
plt.show()
```

```python
sns.histplot(data=video_game_sale, x="Genre")
plt.title("Sales Distribution")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.savefig("VideGameGenreSales.png",dpi=300)
```

```python
sns.pairplot(video_game_sale, hue="Genre")
plt.savefig("VideGameGenreSalesGraphs.png",dpi=300)
plt.show()
```

```python
# Create a KDE plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=video_game_sale, x="Global_Sales", hue="Genre", fill=True, common_norm=False, palette="crest", alpha=0.2)

# Add labels and title
plt.title("Sales based on different Genre")
plt.xlabel("Genre")
plt.ylabel("Sales")
plt.savefig("GlobalsSales.png",dpi=300)
plt.show()
```

```python
plt.grid(True)
plt.scatter(video_game_sale["NA_Sales"], video_game_sale["EU_Sales"])
plt.title("NA_Sales vs EU_Sales")
plt.xlabel("NA_Sales")
plt.ylabel("EU_Sales")
plt.savefig("NA_Sales vs EU_Sales.png",dpi=300)
```

```python
plt.grid(True)

plt.plot(video_game_sale["Rank"], video_game_sale['Global_Sales'])
plt.title("Global_Sales")
plt.xlabel("Rank")
plt.ylabel("Global_Sales")
# Set x-axis for every 10 units
plt.xticks(range(0, max(video_game_sale["Rank"]) + 1, 10))  
plt.savefig("GlobalSalesPlot.png", dpi=300)
plt.show()
```
